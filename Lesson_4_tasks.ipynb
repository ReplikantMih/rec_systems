### В чем принципиальное отличие гибридных рекомендательных систем от коллаборативной филтьтрации?

В том, что коллаборативная фильтрация не учитывает признаковое описание юзеров и товаров. Что имеет такие последствия как:
* Коллаборативная фильтрация хуже работает с холодными юзерами (и товарами).
* Коллаборативная фильтрация в целом имеет предел по качеству ниже (в ряде случаев).
* Гибридные системы адаптивны и более универсальны (показывают хорошее качество в разных условиях), в то время как коллаборативные системы специализируются на dense данных интеракций, а content-based - на разряженных или отсутствующих.

### Приведите 2-3 примера задач, в которых необходимо использовать гибридные системы

* Интернет магазин находится на этапе активного роста и доли постоянных клиентов и новых сопоставимы, задача - выдача рекомендаций "так же вам может понравиться".

* Интернет книжный магазин или стриминговый сервис или провайдер другого развлекательного контента, необходимо сопровождать рекомендациями интересного контента не только клиентов, у которых уже есть данные по интерациям, но и новых клиентов в начале их жизненного цикла.

Возможные проблемы при выводе в продакшен и при обеспечении работоспособности:
* Система сложная, все возможные проблемы возможны в основном по этой причине. Чем сложнее система, тем больше мест, где что-то может пойти не так, тем выше вероятность, что что-то пойдет не так.

##### Какой датасет используют авторы?

1. MovieLens. Датасет фильмов с рейтингами. Признаковое описание фильмов - теги, пользователи явного признакового описания не имеют. Таргет - рейтинг фильма, был преобразован в бинарный формат. Для эксперимента на основе дата-сета были сгенерированы два разных контекста - warm-start и cold-start. Априорно в дата-сете присутствуют только пользователи, которые поставили от 20 оценок, что делает матрицу взаимодействий относительно плотоной.


2. CrossValidated. Датасет вопросов и ответов на тему машинного обучения. Ответ - интеракция, юзер - юзер, вопрос - item. В данном дата-сете и юзеры и вопросы имели явное признаковое описание, матрица интерации намного более разряженная по сравнению с матрицей MovieLens.

##### С какими моделями сравнивают LightFM? Опишите их основные идеи кратко.

1. MF. Матричная факторизаци. Алгоритм на основе только интеракций. Суть модели - представление юзеров и айтемов в виде проекций на латентное признаковое пространство такой размерности, которая позволяет осуществлять операции матричного умножения между матрицей представления пользователей и айтемов. По алгоритму градиентного спуска матрицы описаний обучаются.


2. LSI-LR. Content-based модель. Соответственно, использует только явное признаковое описание пользователей и айтемов, не задействует данные по взаимодействию.


3. LSI-UP. Гибридная модель.
