# Основные задания:

### 1. В чем принципиальные отличия item-item рекомендаций от ALS?

К принципиальным отличия можно отнести, вероятно, следующие:

* Item-item - непредобучаемая модель, ALS - обучаемая.
* Из первого пункта вытекают различия в сложности и, соответственно, скорости предикта, ALS - быстро, item-item - очень медленно. 

### 2. Приведите 3 примера весов user-item матрицы для задачи рекоммендаций товаров.

Если я правильно понял вопрос, имеется в виду: на что можно взвешивать значения ячеек в матрице user-item для корректировки результатов под те или иные задачи.

Варианты весов:

* Взвешивать на некоторую характеристику доходности/маржинальности. Например, товарам с большей маринальностью и/или большим абсолютным значением дохода давать веса больше. Для целей увеличения именно дохода, а не продаж, т.е. затачивание под конкретные экономические метрики.

* Взвешивать под задачи увеличения продаж по конкретным группам, наименованиям товаров и т.д. Например, чтобы избавиться от залежалого товара или чтобы повысить оборот по конкретному поставщику для получения более выгодных условий и т.д.

* Возможны, вероятно, какие-то варианты взвешивания в разрезе пользователей тоже, по какой-то причине мы считаем что какая-то группа пользователей имеет больше вес (более компетентна, например - предположим, у нас технически сложный товар и мы по какой-то метрике можем ранжировать пользователей по условной компетентности и считаем что опираться на их выбор надежней). Например, простейший вариант - взешивать на основе времени жизни аккаунта с идеей, что чем дольше человек с нами, тем болье он в этом разбирается - навернякак, такая корреляция (даже если не сильная) есть.

### 3. Какие ограничения есть у ALS? (Тип информации, линейность/нелинейность факторов и т. д.)

Ограничения ALS:

* Эта модель никак не учитывает явные характеристики пользователей и товаров, так что максимум модель может обучиться до некоторого предела, который будет всегда ниже чем предел в случае использования полного набора информационных срезов.
* Пересечение user-item - это всегда скаляр, таким образом, даже имея богатую информацию по конкреному случаю взаимодействия, мы вынуждены кодировать его в скаляр, что, понятно, приводит к потере информации.
* Модель не умеет эффективно работать с новыми пользователями.

### Сравните метрики hit_rate@k, precision@k. Какую матрику использовать предпочтительно и почему. Привидите пример 2-3 задач (опишите, что является клиентом, что товаром), в которой более уместно использовать метрику hit_rate?

* 
Метрика precision@k несет в себе больше информации, чем hit_rate@k, в большинстве случаев предпочитетельней использовать именно её. По сути hit_rate@k - отражение метрики precision@k в бинарное пространство, т.е. дополнительно ничего не дает, но информацию теряет.



* Примеры задач где hit_rate@k предпочтительнее. Если обобщить - это задачи, где теряемая информация (относительно precision@k) не важна, гораздо важнее оценить сам факт, сработала ли рекомендация или нет. Примеры задач:

* 1. Компания продает мобильные телефоны (покупатель - user, товар - item) и компания продает телефоны с отрицательной или нулевой маржой (или очень низкой), заказ отбивается только если с основным товаром покупатель приобретает хотя бы один аксессуар (с нормальной маржой), если компании хочет минимизировать долю убыточных заказов использование метрики hit_rate@k выглядит оправданным.

* 2. Поисковая выдача в поисковике (пользователь - user, ссылка в выдаче - item), если пользователь нашел нужную ссылку с первого раза - ничуть не хуже, а-то и лучше, чем если он протыкал все ссылки, применение hit_rate@k оправдано.


### В метрике NDCG@k мы используем логарифм в знаменатиле. Как Вы думаете, почему именно логарифм? Какую функцию можно использовать вместо логарифма? Привидите пример метрик/подходов к предобработке данных/функций ошибок в ML, где также в знаменателе присутсвует логарифм.

Используется логарифм как частный случай возрастающей функции - для обеспечения принципа: чем дальше рекомендация в ТОПе рекомендаций, тем меньше влияние на метрику. Так же данная функция характеризуется уменьшающейся скоростью возрастания - что так же оправдано с точки зрения данного принципа поскольку падение "логического значения" рекомендаций при удаленее от ТОП1 так же идет нарастающим темпом (как и 1/log).

Вместо логарифма может быть использована любая другая функция с аналогичными озвученным характеристиками, например Х ** (1/2).
